```python
import requests
from bs4 import BeautifulSoup
ReviewStr = list()
PriceStr=list()
ProductName=list()
# function to scrap data from a given url
def scrappData(url):
    URL=url 
    cnt = requests.get(URL)
    htmlContent=cnt.content
    soup = BeautifulSoup(htmlContent, 'html5lib') 
    script=soup.select_one('script[type="application/ld+json"]').text
    json_obj=json.loads(script)
    reviews_list=json_obj['review']
    for review in reviews_list:
        ReviewStr.append(review['reviewBody'])
    price_list=json_obj["offers"]
    PriceStr.append(price_list['lowPrice'])
#     span=soup.select_one('div[class=""pdp-mod-product-badge-wrapper""data-spm-anchor-id=""a2a0e.pdp.0.i3.3984643dkBWiZM""]').text
#     json_obj=json.loads(span)
#     ProductName.append()    
# given pages url to scrap
page1=scrappData("https://daraz.pk/products/pack-of-5-beotua-plant-flavor-hand-cream-set-moisturizing-nourishing-anti-chapping-i136450553-s1296722160.html?spm=a2a0e.searchlist.list.13.482b60dfz2tGTY&search=1")
page2=scrappData("https://daraz.pk/products/wb-by-hemani-moisturizing-orchid-hand-cream-30ml-i3316390-s12733695.html?spm=a2a0e.searchlist.list.1.482b60dfIYvCq4&search=1")
page3=scrappData("https://daraz.pk/products/set-of-5-plant-extract-fragrance-moisturizing-nourishing-hand-creams-i175236812-s1348096344.html?spm=a2a0e.searchlist.list.3.482b60dfF9F3XN&search=1")
page4=scrappData("https://daraz.pk/products/wb-by-hemani-be-soft-hand-cream-with-goat-milk-75ml-i1364243-s8997813.html?spm=a2a0e.searchlist.list.5.482b60dfaMgmOO&search=1")
page5=scrappData("https://www.daraz.pk/products/pack-of-5-beotua-plant-flavor-hand-cream-set-i136450602-s1296716708.html?spm=a2a0e.searchlist.list.17.482b60dfUhQXCb&search=1")

# join all reviews list to a single string
string_value = ".".join(ReviewStr)
```

```python
# break string into lines and remove extra spaces

seperatedData= re.split('[,.!]', string_value)
seperatedData = [x.strip() for x in seperatedData if x.strip()]

emojiFreeData = list()
# remove emojies from the data  

def replace_emoji_space(string):
    emoji_pattern = re.compile("["
                            "\U0001F1E0-\U0001F1FF"  # flags (iOS)
                            "\U0001F300-\U0001F5FF"  # symbols & pictographs
                            "\U0001F600-\U0001F64F"  # emoticons
                            "\U0001F680-\U0001F6FF"  # transport & map symbols
                            "\U0001F700-\U0001F77F"  # alchemical symbols
                            "\U0001F780-\U0001F7FF"  # Geometric Shapes Extended
                            "\U0001F800-\U0001F8FF"  # Supplemental Arrows-C
                            "\U0001F900-\U0001F9FF"  # Supplemental Symbols and Pictographs
                            "\U0001FA00-\U0001FA6F"  # Chess Symbols
                            "\U0001FA70-\U0001FAFF"  # Symbols and Pictographs Extended-A
                            "\U00002702-\U000027B0"  # Dingbats
                            "\U000024C2-\U0001F251"
                       "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r' ', str(string))
for i in seperatedData:
    emojiFreeData.append(replace_emoji_space(i))
for i in emojiFreeData:
    print(i)
```

```python
ReviewDataFrame = pd.DataFrame(emojiFreeData,columns=['Reviews'])
ReviewDataFrame.head()
```

```python
# fill empty strings with Average
ReviewDataFrame=ReviewDataFrame.replace(r'^\s*$',"Average", regex=True)
# convert to lower case
ReviewDataFrame["Reviews"] = ReviewDataFrame["Reviews"].str.lower()
ReviewDataFrame.head()
```

```python
# load data from file to list
filename1="positiveword.txt"
positveWords =[]
with open(filename1, 'r') as file:
    for line in file:
        positveWords.append(line.strip().split(","))
print(positveWords)
filename2="negativeword.txt"
NegativeWords =[]
with open(filename2, 'r') as file:
    for line in file:
        NegativeWords.append(line.strip().split(","))
print(NegativeWords)
```

```python
ReviewDataFrame_copy = ReviewDataFrame.copy()
finalRemarksList=[None] * len(ReviewDataFrame_copy['Reviews'])
finalRemarksCounter = 0

NegativeWords_flat_list = []
for sublist in NegativeWords:
    for item in sublist:
        NegativeWords_flat_list.append(item)

# print(NegativeWords_flat_list)

for i in ReviewDataFrame_copy['Reviews']:
#     print(i)
    string1_words = set(i.split())
    string2_words = set(NegativeWords_flat_list)
    common_words = string1_words & string2_words
#     print(common_words)
    if(len(common_words)>0):
        print('negative')
        finalRemarksList[finalRemarksCounter]='negative'
        finalRemarksCounter+=1

positveWords_flat_list = []
for sublist in positveWords:
    for item in sublist:
        positveWords_flat_list.append(item)

# print(NegativeWords_flat_list)

for i in ReviewDataFrame_copy['Reviews']:
    print(i)
    string1_words = set(i.split())
    string2_words = set(positveWords_flat_list)
    common_words = string1_words & string2_words
    print(common_words)
    if(len(common_words)>0):
        print('positive')
        finalRemarksList[finalRemarksCounter]='positive'
        finalRemarksCounter+=1

finalRemarksList = ['Average' if v is None else v for v in finalRemarksList]

ReviewDataFrame_copy['Remarks']=finalRemarksList

ReviewDataFrame_copy    
        
        
```
